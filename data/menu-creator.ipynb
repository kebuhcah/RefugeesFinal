```python
from bs4 import BeautifulSoup
```

```python
bs = BeautifulSoup()
table = bs.new_tag('table')
bs.append(table)
table['border'] = "5"
table['cellpadding'] = "5"
for year in range(1999,2016):
    tr = bs.new_tag('tr')
    bs.table.append(tr)
    yearcell = bs.new_tag('td')
    tr.append(yearcell)
    yearcell.append(str(year))
    for sheet in (['SouthWestAsia','Afghanistan', 'Iran', 'Pakistan'] if year < 2014 else ['AsiaPacific']):        
        td = bs.new_tag('td')
        a = bs.new_tag('a')
        tr.append(td)
        td.append(a)
        a.append(sheet)
        a['href'] = 'unhcr_reports/' + str(year) + '-' + sheet + '.pdf'
print bs.prettify()

html = bs.prettify("utf-8")
with open("menu.html", "wb") as file:
    file.write(html)
```
